EDA of Initial Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets
application_train = pd.read_csv(r'C:\Users\HP\Desktop\Credit_risk\application_train.csv')
application_test = pd.read_csv(r'C:\Users\HP\Desktop\Credit_risk\application_test.csv')
bureau = pd.read_csv(r'C:\Users\HP\Desktop\Credit_risk\bureau.csv')
bureau_balance = pd.read_csv(r'C:\Users\HP\Desktop\Credit_risk\bureau_balance.csv')
POS_CASH_balance = pd.read_csv(r'C:\Users\HP\Desktop\Credit_risk\POS_CASH_balance.csv')
credit_card_balance = pd.read_csv(r'C:\Users\HP\Desktop\Credit_risk\credit_card_balance.csv')
previous_application = pd.read_csv(r'C:\Users\HP\Desktop\Credit_risk\previous_application.csv')
installments_payments = pd.read_csv(r'C:\Users\HP\Desktop\Credit_risk\installments_payments.csv')
```

```python
# Display first few rows of each dataset
datasets = {
    'application_train': application_train,
    'application_test': application_test,
    'bureau': bureau,
    'bureau_balance': bureau_balance,
    'POS_CASH_balance': POS_CASH_balance,
    'credit_card_balance': credit_card_balance,
    'previous_application': previous_application,
    'installments_payments': installments_payments
}

for name, data in datasets.items():
    print(f"\n{name}:\n", data.head())
```

```python
# Check shape and summary statistics
for name, data in datasets.items():
    print(f"\n{name} shape: {data.shape}")
    print(f"\n{name} summary statistics:\n", data.describe(include='all'))
```

```python
# Visualize missing values
for name, data in datasets.items():
    plt.figure(figsize=(10, 6))
    sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
    plt.title(f'Missing Values in {name}')
    plt.show()

    print(f"\n{name} missing values:\n", data.isnull().sum())
```

```python
# Target Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='TARGET', data=application_train)
plt.title('Target Distribution')
plt.show()
```

```python
# Histograms for numerical features
num_features = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'DAYS_BIRTH']
n_cols = 2
n_rows = (len(num_features) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
fig.suptitle('Histograms of Numerical Features', fontsize=16)

for i, feature in enumerate(num_features):
    row = i // n_cols
    col = i % n_cols
    ax = axes[row, col]
    application_train[feature].hist(bins=30, ax=ax)
    ax.set_title(feature)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

The income distribution is highly skewed, with most clients earning below 500,000. Credit amounts primarily fall under 1.5 million, showing distinct peaks. Annuities are also right-skewed, mostly below 50,000. Age distribution (in days) is roughly normal, with a concentration around ages 41 to 55. This variability is crucial for assessing creditworthiness.

```python
# Bar plots for categorical features
cat_features = ['NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY']
n_cols = 2
n_rows = (len(cat_features) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
fig.suptitle('Distribution of Categorical Features', fontsize=16)

for i, feature in enumerate(cat_features):
    row = i // n_cols
    col = i % n_cols
    ax = axes[row, col]
    sns.countplot(x=feature, data=application_train, ax=ax)
    ax.set_title(feature)
    ax.tick_params(axis='x', rotation=90)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```


The majority of loans are cash loans, with a smaller proportion being revolving loans. Female applicants significantly outnumber male applicants. Most clients do not own a car, while the majority own real estate. These distributions suggest a client base with more traditional loan types, a higher proportion of females, and a tendency towards real estate ownership over car ownership. This shows the demographics and financial behavior of the client base.

```python
# Select only numerical columns
numerical_columns = application_train.select_dtypes(include=['float64', 'int64']).columns
corr_matrix = application_train[numerical_columns].corr()

# Plot the heatmap without annotations
plt.figure(figsize=(20, 15))
sns.heatmap(corr_matrix, cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

The correlation matrix reveals a cluster of highly correlated features, particularly among the apartment-related variables. Some features exhibit strong negative or positive correlations, indicating potential multicollinearity.

```python
# Create a table for the top features
target_corr = corr_matrix['TARGET'].abs().sort_values(ascending=False)
top_features_corr = pd.DataFrame(target_corr).reset_index()
top_features_corr.columns = ['Feature', 'Correlation with TARGET']

# Display the table of top correlated features
top_features_corr
```

```python
# Box plots for Outlier Detection
num_features = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'DAYS_EMPLOYED']
n_cols = 2
n_rows = (len(num_features) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
fig.suptitle('Box Plots for Numerical Features', fontsize=16)

for i, feature in enumerate(num_features):
    row = i // n_cols
    col = i % n_cols
    ax = axes[row, col]
    sns.boxplot(y=feature, data=application_train, ax=ax)
    ax.set_title(feature)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

The box plots indicate significant outliers in AMT_INCOME_TOTAL, AMT_CREDIT, and DAYS_EMPLOYED. Most values for these features are concentrated near the lower end, with a few extreme values stretching the distribution. This suggests a need for careful handling of outliers in the dataset to improve model performance and interpretation.

```python
# plots for features vs. target
num_features = ['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'DAYS_BIRTH']
n_cols = 2
n_rows = (len(num_features) + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
fig.suptitle('Features vs. Target', fontsize=16)

for i, feature in enumerate(num_features):
    row = i // n_cols
    col = i % n_cols
    ax = axes[row, col]
    sns.boxplot(x='TARGET', y=feature, data=application_train, ax=ax)
    ax.set_title(f'{feature} vs. TARGET')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

AMT_CREDIT shows a slight difference in distribution between the target classes, indicating it may have some influence on loan repayment outcomes.

```python
# Count of previous applications per client
prev_app_count = previous_application['SK_ID_CURR'].value_counts()
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

# Plot 1
sns.histplot(prev_app_count, bins=50, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Previous Applications per Client')
axes[0].set_xlabel('Number of Previous Applications')
axes[0].set_ylabel('Number of Clients')

# Plot 2
sns.histplot(previous_application['AMT_CREDIT'], bins=50, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Loan Amounts in Previous Applications')
axes[1].set_xlabel('Loan Amount')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Most clients have only a few previous applications, with the number decreasing sharply as it increases. Loan amounts are highly skewed, with most being relatively low.

```python
# Distribution of credit types in the bureau dataset
plt.figure(figsize=(12, 6))
sns.countplot(y='CREDIT_TYPE', data=bureau, order=bureau['CREDIT_TYPE'].value_counts().index)
plt.title('Distribution of Credit Types in Bureau Dataset')
plt.xlabel('Count')
plt.ylabel('Credit Type')
plt.show()

# Create a plot for status counts over time
status_counts = bureau_balance.groupby('MONTHS_BALANCE')['STATUS'].value_counts().unstack().fillna(0)
plt.figure(figsize=(15, 6))
status_counts.plot(kind='line', stacked=True, colormap='viridis', ax=plt.gca())
plt.title('Credit Status Over Time in Bureau Balance Dataset')
plt.xlabel('Months Balance')
plt.ylabel('Count of Status')
plt.legend(title='Credit Status')
plt.show()
```

The majority of credits in the bureau dataset are consumer credits, followed by credit card loans. The credit status over time shows that 'X' and 'C' statuses have the highest counts, indicating many loans were closed or had no payment required.

```python
# Number of payments per loan
payments_per_loan = installments_payments['SK_ID_PREV'].value_counts()
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

sns.histplot(payments_per_loan, bins=50, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Number of Payments per Loan')
axes[0].set_xlabel('Number of Payments')
axes[0].set_ylabel('Frequency')

sns.histplot(installments_payments['AMT_PAYMENT'], bins=50, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Amounts Paid in Installments')
axes[1].set_xlabel('Amount Paid')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

The distribution of the number of payments per loan shows that most loans have relatively few payments, with a steep drop-off as the number of payments increases. The amounts paid in installments are highly skewed, with most payments being relatively small. This indicates that clients typically repay loans with fewer, smaller payments.

```python
# Convert days to years for easier interpretation
application_train['YEARS_BIRTH'] = application_train['DAYS_BIRTH'] / -365
application_train['YEARS_EMPLOYED'] = application_train['DAYS_EMPLOYED'] / -365
fig, axes = plt.subplots(1, 2, figsize=(20, 6))

sns.histplot(application_train['YEARS_BIRTH'], bins=50, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Client Age')
axes[0].set_xlabel('Age (Years)')
axes[0].set_ylabel('Frequency')

sns.histplot(application_train['YEARS_EMPLOYED'], bins=50, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Employment Duration')
axes[1].set_xlabel('Years Employed')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

The client age distribution is relatively uniform between 30 and 60 years, indicating a broad age range of borrowers. The employment duration distribution shows a high concentration around zero, with an outlier suggesting possible data errors.

Our initial EDA reveals that most clients have limited previous loan applications and tend to borrow smaller amounts. The data shows significant outliers in income, credit, and employment duration. Key features such as credit type distribution highlight the predominance of consumer credit and varied credit statuses over time. The client demographic is broadly aged between 30 and 60 years, with employment data suggesting potential data quality issues.

EDA of Cleaned Dataset (Preprocess + Feature engineering)

```python
train_path = r'C:\Users\HP\Desktop\Credit_risk\250 features\final_application_train.csv'
test_path = r'C:\Users\HP\Desktop\Credit_risk\250 features\final_application_test.csv'

application_train = pd.read_csv(train_path)
application_test = pd.read_csv(test_path)
```

```python
# Display the column names
train_columns = application_train.columns.tolist()
test_columns = application_test.columns.tolist()

print("\nTraining Data Columns:\n", train_columns)
print("\nTest Data Columns:\n", test_columns)
```

```python
# Summary statistics
train_summary = application_train.describe()
test_summary = application_test.describe()

print("\nTraining Data Summary:\n", train_summary)
print("\nTest Data Summary:\n", test_summary)
```

```python
# Plot histograms and KDE plots
numerical_features = [
    'EXT_SOURCE_MEAN', 'CREDIT_INCOME_RATIO',
    'DAYS_BIRTH', 'DAYS_EMPLOYED', 'AMT_CREDIT', 'AMT_INCOME_TOTAL'
]

fig, axes = plt.subplots(len(numerical_features), 2, figsize=(15, 5 * len(numerical_features)))
for i, feature in enumerate(numerical_features):
    sns.histplot(application_train[feature], bins=30, kde=True, ax=axes[i, 0])
    axes[i, 0].set_title(f'Histogram of {feature}')
    
    sns.kdeplot(application_train[feature], fill=True, ax=axes[i, 1])
    axes[i, 1].set_title(f'KDE of {feature}')

plt.tight_layout()
plt.show()
```

```python
# summary statistics for the aggregated features
agg_features_train = application_train.filter(regex='(BUREAU|POS_CASH|CREDIT_CARD|PREV_APP|INSTALLMENTS)')
agg_features_test = application_test.filter(regex='(BUREAU|POS_CASH|CREDIT_CARD|PREV_APP|INSTALLMENTS)')

print("Aggregated Numerical Features - Training Data Summary:")
print(agg_features_train.describe())

print("\nAggregated Numerical Features - Test Data Summary:")
print(agg_features_test.describe())
```

```python
# Select a subset of aggregated features to plot
selected_agg_features = agg_features_train.columns[:6]

fig, axes = plt.subplots(len(selected_agg_features), 1, figsize=(10, 5 * len(selected_agg_features)))

for i, feature in enumerate(selected_agg_features):
    sns.histplot(agg_features_train[feature], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```
